```python
from selenium import webdriver
from selenium.webdriver.firefox.service import Service
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.firefox.firefox_profile import FirefoxProfile
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from webdriver_manager.firefox import GeckoDriverManager
import time
```

```python
profile_path = r"/Users/gloria/Library/Application Support/Firefox/Profiles/4czc69sv.default-release" 
profile = FirefoxProfile(profile_path)

options = Options()
options.profile = profile
options.add_argument("--width=1400")
options.add_argument("--height=900")

driver = webdriver.Firefox(
    service=Service(GeckoDriverManager().install()),
    options=options
)
```

```python
driver.get("https://x.com/search?q=true%20OR%20real%20(to%3AWhipKClark)%20(%40grok)%20lang%3Aen&src=typed_query&f=top")
```

```python

SCROLL_PAUSE = 2  
DETAIL_PAUSE = 3  

last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    driver.find_element(By.TAG_NAME, "body").send_keys(Keys.END)
    time.sleep(SCROLL_PAUSE)
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break  
    last_height = new_height

print(" 页面滚动完成，所有推文已加载到 DOM")
```

```python
idx = 0
while True:
    all_articles = driver.find_elements(By.TAG_NAME, "article")
    
    if idx >= len(all_articles):
        print(" 所有推文已处理完成 ")
        break

    try:
        article = all_articles[idx]

        driver.execute_script(
            "arguments[0].scrollIntoView({behavior:'smooth', block:'center'});",
            article
        )
        time.sleep(1)

        article.click()
        print(f" 已进入第 {idx+1} 条推文页面")
        time.sleep(DETAIL_PAUSE)  

        driver.back()
        print(f" 已返回列表页")
        time.sleep(2)

        idx += 1 

    except Exception as e:
        print(f" 跳过第 {idx+1} 条推文（错误）：{e}")
        try:
            driver.back()
        except:
            pass
        time.sleep(2)
        idx += 1
```
